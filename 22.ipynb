```python
from data_loader import DataLoader, DataVisualizer, DataCleaner

# Загружаем данные из CSV файла
data = DataLoader.load_csv('archive.csv')

# Выводим первые строки и названия колонок для проверки
print(data.head())
print(data.columns)

# Создаем объект для визуализации данных
visualizer = DataVisualizer()

# Обновляем имена колонок в соответствии с теми, что выведены выше
# Гистограмма выбросов CO2
visualizer.add_histogram(data, 'Carbon Dioxide (ppm)', title="Гистограмма выбросов CO2")

# Линейный график выбросов CO2 по годам
visualizer.add_line_plot(data, 'Decimal Date', 'Carbon Dioxide (ppm)', title="Выбросы CO2 по годам")

# Диаграмма рассеяния для CO2 и сезонно скорректированного CO2
visualizer.add_scatter_plot(data, 'Carbon Dioxide (ppm)', 'Seasonally Adjusted CO2 (ppm)', title="Выбросы CO2 в сравнении с Seasonally Adjusted CO2 (ppm)")

# Отображаем все визуализации
visualizer.show_visualizations()

# Создаем объект для очистки данных
cleaner = DataCleaner()

# Считаем и выводим количество пропущенных значений
missing_values_report = cleaner.report_missing_values(data)
print(missing_values_report)

# Заполняем пропущенные значения средним
cleaned_data = cleaner.fill_missing_values(data, method='mean')
print(cleaned_data.head())
```
